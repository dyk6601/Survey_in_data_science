# Problem 2

```python
# 2024 season total points
player_points_totals = df.groupby('athlete_display_name')['points'].sum()
games_played = df.groupby('athlete_display_name').size()

players = ["Caitlin Clark", "Angel Reese", "A'ja Wilson"]
for player in players:
    if player in player_points_totals.index:
        total_points = player_points_totals.loc[player]
        games = games_played.loc[player]
        print(f"{player} scored {total_points} total points in {games} games.")
    else:
        print(f"{player} not found in the dataset.")
```

# Problem 3
Compute the average number of points per game played for Clark, Reese, and
Wilson. Specifically, compute the average number of points per game in which
the player played more than 0 minutes. The variable “minutes” records how
many minutes the player played in that game.

```python
# Problem 3: Average points per game (only games where player played > 0 minutes)
games_with_minutes = df[df['minutes'] > 0]  # Filter games where player actually played
player_avg_points_per_game = games_with_minutes.groupby('athlete_display_name')['points'].mean()

players = ["Caitlin Clark", "Angel Reese", "A'ja Wilson"]
for player in players:
    if player in player_avg_points_per_game.index:
        avg_ppg = player_avg_points_per_game.loc[player]
        print(f"{player} has an average of {avg_ppg:.2f} points per game.")
    else:
        print(f"{player} not found in the dataset.")
```

```python
# find their top two games by number of points scored. Report the numbers of points, dates, and opponents for those two games.
players = ["Caitlin Clark", "Angel Reese", "A'ja Wilson"]
player_top_2_point_games = df.sort_values('points',ascending=False).groupby('athlete_display_name').head(2)
for player in players:
    player_games = player_top_2_point_games[player_top_2_point_games['athlete_display_name'] == player]
    print(f"Top 2 scoring games for {player}:")
    print(player_games[['points', 'game_date', 'opponent_team_name']])
```
